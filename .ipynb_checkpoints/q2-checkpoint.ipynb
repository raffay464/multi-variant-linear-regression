# Generating Dataset

To generate dataset, ill fill in the variables (both independent) with random values. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
def generate_dataset(num_entries=20):
    # Seed for reproducibility
    np.random.seed(0)

    # Independent variables
    sleep_hours = np.random.randint(1, 11, size=num_entries)
    rank_of_food_quality = np.random.randint(1, 11, size=num_entries)
    working_hours = np.random.randint(1, 11, size=num_entries)
    rank_of_health_condition = np.random.randint(1, 11, size=num_entries)
    rank_of_mood = np.random.randint(1, 11, size=num_entries)

    rank_of_productivity = (
        2 * sleep_hours +
        3 * rank_of_food_quality -
        1 * working_hours +
        2 * rank_of_health_condition +
        2 * rank_of_mood +
        np.random.randint(-2, 3, size=num_entries)  # Noise
    )

    # Ensure the dependent variable is within the 1-10 range
    rank_of_productivity = np.clip(rank_of_productivity, 1, 10)

    # Creating a DataFrame
    data = pd.DataFrame({
        'Sleep Hours': sleep_hours,
        'Rank of Food Quality': rank_of_food_quality,
        'Working Hours': working_hours,
        'Rank of Health Condition': rank_of_health_condition,
        'Rank of Mood': rank_of_mood,
        'Rank of Productivity': rank_of_productivity
    })

    return data

# Generate the dataset
data = generate_dataset()

# Save the DataFrame to a CSV file
csv_file_path = 'generated_data.csv'
data.to_csv(csv_file_path, index=False)

csv_file_path 
```



# Implementing Gradient Descent 

```python
data = pd.read_csv(csv_file_path)

# Define the cost function
def compute_cost(X, y, theta):
    m = len(y)  
    predictions = X.dot(theta)
    cost = (1/(2*m)) * np.sum(np.square(predictions - y))
    return cost

# Define the gradient descent function
def gradient_descent(X, y, theta, learning_rate, num_epochs):
    
    
    m = len(y)
    cost_history = np.zeros(num_epochs)
    
    for epoch in range(num_epochs):
        gradients = (1/m) * X.T.dot(X.dot(theta) - y)
        theta = theta - learning_rate * gradients
        cost_history[epoch] = compute_cost(X, y, theta)
        
    return theta, cost_history

# Prepare the data
m, n = data.shape
X = data.iloc[:, :n-1] 
X = np.column_stack((np.ones(m), X)) 
y = data.iloc[:, -1] 
theta_initial = np.random.rand(n)  

# Hyperparameters
learning_rate = 0.01
num_epochs = 1000

# Run gradient descent
theta_final, cost_history = gradient_descent(X, y, theta_initial, learning_rate, num_epochs)

# The final theta values and the last 10 cost values
theta_final, cost_history[-10:]
```

# Gradient Descent

```python
# Define the model hypothesis function: h(x) = mx + c
def hypothesis(X, theta):
    return np.dot(X, theta)

# Define the cost function
def compute_cost(X, y, theta):
    m = len(y)
    h = hypothesis(X, theta)
    return (1/(2*m)) * np.sum((h - y)**2)

# Perform gradient descent to learn theta
def gradient_descent(X, y, theta, learning_rate, num_epochs):
    m = len(y)
    cost_history = []
    theta_history = []

    for epoch in range(num_epochs):
        # Calculate the gradient
        gradient = (1/m) * np.dot(X.T, (hypothesis(X, theta) - y))
        # Update the parameters
        theta = theta - learning_rate * gradient
        
        # Save the cost for plotting
        cost = compute_cost(X, y, theta)
        cost_history.append(cost)
        theta_history.append(theta)

    return theta, cost_history, theta_history

# Prepare the data
X = data.iloc[:, :-1].values  # Independent variables
X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
y = data.iloc[:, -1].values  # Target variable

# Initialize theta
theta_initial = np.random.rand(X.shape[1])

# Set the hyperparameters
learning_rate = 0.01

# Run gradient descent
theta, cost_history, theta_history = gradient_descent(X, y, theta_initial, learning_rate, no_of_epochs)

# Plotting
#cost history over iterations
plt.plot(cost_history)
plt.xlabel('Epoch')
plt.ylabel('Cost')
plt.title('Cost History')
plt.show()

# Output the final theta values
theta
```

# Plotting the accuracies

```python
#to calculate the accuracies we will first find the r-Squared Error and for this we will first create the func for this
# Function to calculate R-squared
def r_squared(y, y_pred):
    ss_res = np.sum((y - y_pred) ** 2)  # Sum of squares of residuals
    ss_tot = np.sum((y - np.mean(y)) ** 2)  # Total sum of squares
    return 1 - (ss_res / ss_tot)

# Calculating R-squared for each epoch
r_squared_history = [r_squared(y, hypothesis(X, theta)) for theta in theta_history]

# Plotting the R-squared values over epochs
plt.figure(figsize=(10, 6))
plt.plot(r_squared_history, label='R-squared')
plt.title('R-squared over Epochs')
plt.xlabel('Epoch')
plt.ylabel('R-squared')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D


import numpy as np
import pandas as pd

np.random.seed(0)  # for reproducibility

num_days = 20

sleep_hours = np.random.randint(1, 10, num_days)  # between 1 to 10 hours
food_quality_rank = np.random.randint(1, 10, num_days)  # rank between 1 to 10
working_hours = np.random.randint(1, 10, num_days)  # between 1 to 10 hours
health_condition_rank = np.random.randint(1, 10, num_days)  # rank between 1 to 10
mood_rank = np.random.randint(1, 10, num_days)  # rank between 1 to 10

noise = np.random.normal(0, 1, num_days)  # random noise
productivity = 0.5 * sleep_hours + 0.2 * food_quality_rank + 0.3 * working_hours + 0.1 * health_condition_rank + 0.4 * mood_rank + noise

# Creating a dictionary for our dataset
data = {
    'Sleep Hours': sleep_hours,
    'Food Quality Rank': food_quality_rank,
    'Working Hours': working_hours,
    'Health Condition Rank': health_condition_rank,
    'Mood Rank': mood_rank,
    'Productivity': productivity
}
df = pd.DataFrame(data)
df.head(20)


import random

class LinearRegression:
    def __init__(self, learning_rate=0.01, min_epochs=100, max_epochs=1000, num_features=None):
        self.learning_rate = learning_rate
        self.num_epochs = random.randint(min_epochs, max_epochs)  # Randomly initialize within the specified range
        self.num_features = num_features
        self.m = np.random.randn(num_features, 1)  # Initialize m randomly with shape (num_features, 1)
        self.c = np.random.randn(1)  # Initialize c randomly




    def compute_cost(self, X, y):
        num_samples = len(y)
        predictions = np.dot(X, self.m) + self.c
        cost = (1/(2*num_samples)) * np.sum((predictions - y)**2)
        return cost

    def gradient_descent(self, X, y):
        num_samples = len(y)
        for epoch in range(self.num_epochs):
            # Compute predictions
            predictions = np.dot(X, self.m) + self.c
            
            # Compute gradients
            dm = (1/num_samples) * np.dot(X.T, (predictions - y))
            dc = (1/num_samples) * np.sum(predictions - y)
            
            # Update parameters
            self.m -= self.learning_rate * dm
            self.c -= self.learning_rate * dc
            
            # Compute and print cost
            cost = self.compute_cost(X, y)
            print(f"Epoch {epoch+1}/{self.num_epochs}, Cost: {cost}")

    def fit(self, X, y):
        self.gradient_descent(X, y)

    def predict(self, X):
        return np.dot(X, self.m) + self.c

# Extracting independent and target variables
X = df[['Sleep Hours', 'Food Quality Rank', 'Working Hours', 'Health Condition Rank', 'Mood Rank']].values
y = df['Productivity'].values.reshape(-1, 1)


# Scatter plot for original data
plt.figure(figsize=(12, 8))
sns.pairplot(pd.DataFrame(data))
plt.show()

# Correlation of independent variables with target variable
correlation_matrix = pd.DataFrame(data).corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Extracting independent and target variables
X = df[['Sleep Hours', 'Food Quality Rank', 'Working Hours', 'Health Condition Rank', 'Mood Rank']].values
y = df['Productivity'].values.reshape(-1, 1)

# Adding bias term
X_bias = np.c_[np.ones((X.shape[0], 1)), X]

# Training the model
lr_model = LinearRegression(min_epochs=500, max_epochs=1500, num_features=X_bias.shape[1])
lr_model.fit(X_bias, y)

# Plotting the best fit line at each epoch
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

for epoch in range(lr_model.num_epochs):
    y_pred = np.dot(X_bias, lr_model.m) + lr_model.c
    ax.scatter(X[:, 0], X[:, 1], y, color='blue')
    ax.plot_trisurf(X[:, 0], X[:, 1], y_pred.flatten(), alpha=0.5, color='red')
    ax.set_xlabel('Sleep Hours')
    #ax.set_ylabel('Food Quality Rank')
    ax.set_zlabel('Productivity')
    ax.set_title(f'Epoch {epoch+1}/{lr_model.num_epochs}')
    plt.pause(0.1)
    ax.clear()

# Plotting the working of gradient descent
plt.figure(figsize=(12, 8))
plt.plot(range(lr_model.num_epochs), cost)
plt.xlabel('Epoch')
plt.ylabel('Cost')
plt.title('Gradient Descent: Epoch vs Cost')
plt.show()
```
